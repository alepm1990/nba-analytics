```python
#Importar las librerías
import pandas as pd
```

```python
#Cargar el dataset con stats NBA

url = "https://www.basketball-reference.com/leagues/NBA_2024_per_game.html"

# Extraer todas las tablas de la página
tables = pd.read_html(url)

# Ver cuántas tablas se extrajeron
print(len(tables))

# Seleccionar la tabla correcta (normalmente la primera)
df = tables[0]
```

```python
df.head(10)
```

```python
# Análisis previo del dataset
df.info()
df.describe()
```

```python
# Eliminar filas y columnas innecesarias
df = df.drop(columns=['Rk', 'Awards'])
```

```python
df = df.drop(df.index[-1])
```

```python
df = df[(df['Team'] != '2TM') & (df['Team'] != '3TM')]
```

```python
#Consutar si existen valores nulos
df.isnull().sum()
```

```python
# Sustituir los valores nulos
valores_nuevos = {"3P%": 0, "2P%":0, "eFG%":0, "FT%":0, "FG%":0}
df = df.fillna(valores_nuevos)

# Lista de columnas que queremos consultar
columns_to_check = ['3P%', '2P%', 'eFG%', 'FT%', 'FG%']

# Contar valores nulos en esas columnas
df[columns_to_check].isnull().sum()
```

```python
df.head(10)
```

```python
#Añadimos nuevas columnas con información relevante para el análisis
df['PTS_por_minuto'] = (df['PTS'] / df['MP']).round(2)
df['triples_lanzados'] = (df['3PA'] / df['FGA']).round(2)
df['lanzamientos_dos'] = (df['2PA'] / df['FGA']).round(2)
df.fillna(0, inplace=True)
df.head(7)
```

Consulta de estadísticas por equipo / jugador

```python
# Top 10 jugadores con más puntos por partido
top_scorers = df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(10)
print(top_scorers)
```

```python
# Top 5 equipos con mayor anotación promedio
top_team_scorers = df.groupby('Team')['PTS'].sum().sort_values(ascending=False).head(5)
print(top_team_scorers)
```

```python
# Top 10 equipos con mejor porcentaje de 3 puntos
team_stats = df.groupby('Team')['3P%'].mean().sort_values(ascending=False)
print(team_stats.head(10))
```

```python
# Definición variable 'Equipos'
equipos = df['Team'].unique()
equipos
```

```python
# Líderes estadísticos por equipo
print("\nLíderes por equipo")
print("\nSelecciona una estadística:")
print("PTS (puntos)", "AST (asistencias)", "TRB (rebotes)", "STL (robos)", "BLK (tapones)")

estadistica = input("Estadística a consultar").lower()
if estadistica == "pts" or estadistica == "puntos":
    for e in equipos:
        max_fila_pts = df[df['Team'] == e].loc[df[df['Team'] == e]['PTS'].idxmax()]
        print(e, max_fila_pts['Player'], max_fila_pts['PTS'])
elif estadistica == "ast" or estadistica == "asistencias":
    for e in equipos:
        max_fila_asist = df[df['Team'] == e].loc[df[df['Team'] == e]['AST'].idxmax()]
        print(e, max_fila_asist['Player'], max_fila_asist['AST'])
elif estadistica == "trb" or estadistica == "rebotes":
    for e in equipos:
        max_fila_reb = df[df['Team'] == e].loc[df[df['Team'] == e]['TRB'].idxmax()]
        print(e, max_fila_reb['Player'], max_fila_reb['TRB'])
elif estadistica == "stl" or estadistica == "robos":
    for e in equipos:
        max_fila_stl = df[df['Team'] == e].loc[df[df['Team'] == e]['STL'].idxmax()]
        print(e, max_fila_stl['Player'], max_fila_stl['STL'])
elif estadistica == "blk" or estadistica == "tapones":
    for e in equipos:
        max_fila_blk = df[df['Team'] == e].loc[df[df['Team'] == e]['BLK'].idxmax()]
        print(e, max_fila_blk['Player'], max_fila_blk['BLK'])
else:
    print("No has escrito una estadística válida, inténtelo otra vez")
```

```python
# Consulta de líderes estadísticos por equipo

for n, e in enumerate(equipos, 1):
    print(f"{n} - {e}")
seleccion_equipo = input("Selecciona un equipo (siglas del listado):").upper()
equipo = df[df['Team'] == seleccion_equipo]

jugador_pts = equipo.sort_values(by='PTS', ascending=False).iloc[0]
jugador_ast = equipo.sort_values(by='AST', ascending=False).iloc[0]
jugador_reb = equipo.sort_values(by='TRB', ascending=False).iloc[0]
jugador_blk = equipo.sort_values(by='BLK', ascending=False).iloc[0]
jugador_stl = equipo.sort_values(by='STL', ascending=False).iloc[0]

print(f"{jugador_pts['Player']}, {jugador_pts['PTS']} puntos")
print(f"{jugador_ast['Player']}, {jugador_ast['AST']} asistencias")
print(f"{jugador_reb['Player']}, {jugador_reb['TRB']} rebotes")
print(f"{jugador_stl['Player']}, {jugador_stl['STL']} robos")
print(f"{jugador_blk['Player']}, {jugador_blk['BLK']} tapones")
```

```python
# Loop para consultar estadísiticas individuales
# Definición de la variable

def pedir_equipo():
    for n, e in enumerate(equipos, 1):
        print(f"{n} - {e}")
    
    seleccion_equipo = input("Selecciona un equipo (siglas del listado): ").upper()
    equipo_2 = df[df['Team'] == seleccion_equipo]

    # Filtrar jugadores con al menos 15 partidos jugados
    equipo_filtrado = equipo_2[equipo_2['G'] >= 10]

    # Crear un diccionario con índices desde 1
    diccionario = {i: player for i, player in enumerate(equipo_filtrado['Player'], 1)}

    print("\nLista de jugadores disponibles:")
    for key, value in diccionario.items():
        print(f"{key} - {value}")

    while True:  # Loop para consultar jugadores
        pregunta = input("\n¿Quieres consultar las estadísticas de un jugador? (si/no): ").lower()
        
        if pregunta == "no":
            print("¡Gracias!")
            break  # Sale del loop si el usuario no quiere continuar

        elif pregunta == "si":
            try:
                indice = int(input("Indica el jugador a consultar escribiendo su valor numérico: "))
                
                if indice in diccionario:
                    jugador_a_consultar = diccionario[indice]
                    df_jugador_filtrado = df[df['Player'] == jugador_a_consultar]
                    print(df_jugador_filtrado) # Mostrar estadísticas del jugador
                else:
                    print("Número inválido, selecciona un valor de la lista.")

            except ValueError:
                print("Por favor, ingresa un número válido.")

        else:
            print("Respuesta no válida. Escribe 'si' o 'no'.")
```

```python
pedir_equipo()
```
